Similar to standard digital computers (classical computers) where the fundamental unit is the bit, quantum computers utilize qubits, which are an extension of the bit into the realm of quantum mechanics.

A more precise definition of a qubit is

<div class="alert alert-block alert-success">
A qubit is a quantum mechanical system with two states (or two levels) that belong to a two-dimensional Hilbert space ($\mathbb{C}^2$). Each qubit can exist in a superposition of both states simultaneously, which is a fundamental concept in quantum computing. The state of a qubit can be represented by a complex vector in the two-dimensional Hilbert space, where each basis vector corresponds to one of the two possible states (typically denoted as $|0\rangle$ and $|1\rangle$). The superposition of these basis states allows qubits to perform computations in parallel and exhibit quantum phenomena such as entanglement and interference.
In quantum computing, qubits are manipulated using quantum gates to perform operations that exploit quantum mechanical principles. The ability of qubits to exist in superposition and entangle with each other forms the basis for the potential computational power of quantum computers compared to classical computers.



A Hilbert space is a vector space equipped with an inner product. The vector space used to define qubits employs the field of complex numbers $\mathbb{C}$ and, as we desire it to be two-dimensional, it is expressed as $\mathbb{C}^2$, meaning we need 2 complex numbers to describe a qubit. To mathematically represent qubits, we use the Dirac notation, also known as bra-ket notation.

The state $\ket{0}$ is pronounced "ket zero", and the state $\ket{1}$ is "ket one". These two states (or qubits) are orthogonal to each other and form a basis $\{\ket{0}, \ket{1}\}$, called the computational basis, because they span the two-dimensional linear vector space (Hilbert space, $\mathbb{C}^2$) of the qubit.

Dirac notation provides a concise way to write vectors. We have mentioned that a qubit is a two-level state, and to represent how much we have of each level, we need a two-dimensional vector. The most general way to write a quantum state (a qubit) is then:

We must have a place to use qubits, and for that, we have the quantum circuit.

A quantum circuit is a model of quantum computation, similar to classical circuits, where computation is performed as a sequence of quantum gates, measurements, qubit initializations to known values, and possibly other actions.

The minimum set of actions that a circuit must be able to perform on qubits to enable quantum computation is known as the DiVincenzo criteria.

These circuits are written in a way where the horizontal axis represents time, starting from the left and ending on the right. The horizontal lines represent qubits, and the double lines represent classical bits. The elements connected by these lines are operations performed on the qubits, such as measurements or gates.
A **qubit** is a fundamental unit of quantum information, analogous to a classical bit but capable of existing in a superposition of states. Mathematically, a qubit state can be represented as:
\[ \alpha \vert 0 \rangle + \beta \vert 1 \rangle \]
where \( \alpha \) and \( \beta \) are complex probability amplitudes such that \( |\alpha|^2 + |\beta|^2 = 1 \), and \( \vert 0 \rangle \) and \( \vert 1 \rangle \) are the computational basis states.

A **quantum circuit** is a model of computation in quantum computing, where operations are performed on qubits using quantum gates.

```python
from qiskit import QuantumCircuit

# Create a quantum circuit with 3 qubits and 2 classical bits
circ = QuantumCircuit(3, 2)

# Apply a Hadamard gate (H) to qubit 0
circ.h(0)

# Apply a CNOT gate (CX) with control qubit 0 and target qubit 1
circ.cx(0, 1)

# Apply another CNOT gate (CX) with control qubit 0 and target qubit 2
circ.cx(0, 2)

# Measure qubit 0 into classical bit 1
circ.measure(0, 1)

# Measure qubit 2 into classical bit 0
circ.measure(2, 0)

# Draw the quantum circuit
print("Quantum Circuit:")
print(circ.draw())

```

The Y gate is a fundamental quantum gate that applies a specific transformation to qubits in a quantum circuit. It is part of the Pauli group of gates, along with the X and Z gates, and it is characterized by its matrix representation:
\textbf{Matrix Representation:}
$$Y = \begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix}$$
The action of the Y gate on the computational basis states is defined as follows
\textbf{The Action of the Y Gate on the Computational Basis States:}
\begin{align*}
Y\ket{0} &= i\ket{1} \\
Y\ket{1} &= -i\ket{0}
\end{align*}
Effect of Y Gate on Superposition States:
This transformation demonstrates how the Y gate affects superposition states, resulting in a phase shift and a change in the relative amplitudes of the basis states.
\textbf{Effect of Y Gate on Superposition States:}
\begin{align*}
Y\ket{+} &= Y\left(\frac{1}{\sqrt{2}}(\ket{0} + \ket{1})\right) \\
&= \frac{1}{\sqrt{2}}(Y\ket{0} + Y\ket{1}) \\
&= \frac{1}{\sqrt{2}}(i\ket{1} - i\ket{0}) \\
&= \frac{-i}{\sqrt{2}}(\ket{0} - \ket{1}) \\
&= -i\ket{-}
\end{align*}


The Y gate is particularly important in quantum computing and quantum algorithms, contributing unique transformations and operations that are essential for manipulating and processing quantum information.

```python
from qiskit.providers.basic_provider import BasicProvider
from qiskit import transpile, QuantumCircuit
from qiskit.visualization import plot_distribution
import numpy as np

# Define the oracle function to mark the item
def oracle(item, n):
    """Returns a quantum circuit implementing the oracle for the specified item."""
    oracle_circuit = QuantumCircuit(n, name='Oracle')

    # Flip qubits where item bit is 0
    for i in range(n):
        if (item >> i) & 1 == 0:
            oracle_circuit.x(i)

    # Multi-controlled Z gate
    oracle_circuit.h(n - 1)
    oracle_circuit.mcx(list(range(n - 1)), n - 1)
    oracle_circuit.h(n - 1)

    # Unflip qubits where item bit is 0
    for i in range(n):
        if (item >> i) & 1 == 0:
            oracle_circuit.x(i)

    return oracle_circuit

# Implement Grover's search algorithm
def grover_search(n, marked_item):
    """Implement Grover's search algorithm to find the marked item in an unsorted list."""
    num_qubits = n
    qc = QuantumCircuit(num_qubits, num_qubits)

    # Apply Hadamard gates to create superposition
    qc.h(range(num_qubits))

    # Optimal number of iterations
    num_iterations = max(1, int(np.pi / 4 * np.sqrt(2 ** num_qubits)))

    # Apply Grover's iterations
    for _ in range(num_iterations):
        # Oracle
        qc.append(oracle(marked_item, num_qubits), range(num_qubits))

        # Diffusion operator
        qc.h(range(num_qubits))
        qc.x(range(num_qubits))
        qc.h(num_qubits - 1)
        qc.mcx(list(range(num_qubits - 1)), num_qubits - 1)
        qc.h(num_qubits - 1)
        qc.x(range(num_qubits))
        qc.h(range(num_qubits))

    # Measure
    qc.measure(range(num_qubits), range(num_qubits))

    return qc

# Please dont change the code below it will be used for grading your circuit
def grader_grover_search(marked_item, n):
    """Grader function to evaluate the Grover's search algorithm."""
    grover_circuit = grover_search(n, marked_item)

    provider = BasicProvider()
    backend = provider.get_backend("basic_simulator")
    new_circuit = transpile(grover_circuit, backend)
    job = backend.run(new_circuit)
    result = job.result().get_counts()

    counts1 = {'00': 525, '11': 499}
    counts2 = {'00': 511, '11': 514}
    legend = ['First execution', 'Second execution']
    plot_distribution([counts1, counts2], legend=legend, color=['crimson','midnightblue'],
                    title="New Distribution")

    # Check if the marked item is identified with high probability
    marked_binary = format(marked_item, f'0{n}b')[::-1]  # Qiskit bit ordering
    if marked_binary in result:
        probability = result[marked_binary] / sum(result.values())
        if probability > 0.8:
            return True
    return False

marked_item = 7
n = 3
grader_grover_search(marked_item, n)

grader_result = grader_grover_search(7, 3)
print("Grader Result:", grader_result)
```

## Exercise 13: *Quantum fourier Transform*

The goal of this exercise is to implement the Quantum Fourier Transform (QFT) on a given quantum circuit using Qiskit. After implementing QFT, we'll use a grading function to verify whether the QFT has been correctly applied to the circuit.

**Tips and Hints:**

*Understanding QFT:* Quantum Fourier Transform (QFT) is a quantum algorithm used for the transformation of quantum states. It's analogous to the classical discrete Fourier transform but operates on quantum states represented by qubits.

*QFT Implementation:* Qiskit provides a built-in library for the Quantum Fourier Transform (QFT), which simplifies the implementation process. We'll use this library to construct the QFT circuit.

*Applying QFT:* To apply QFT to a quantum circuit, use the QFT function from the Qiskit circuit library. This function takes parameters such as the number of qubits (n) and options like do_swaps and inverse to control the behavior of the QFT circuit.

*Grading Function:* After applying QFT to the quantum circuit, we'll compare the resulting circuit with an expected circuit (obtained by applying QFT to an empty circuit). This comparison will ensure that QFT has been correctly implemented.


```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import QFT

# Function to apply Quantum Fourier Transform (QFT) on a quantum circuit
def apply_qft(qc, n):
    qft_circuit = QFT(n, do_swaps=False, inverse=False)  # Create QFT circuit for n qubits
    qc.append(qft_circuit, range(n))  # Append QFT circuit to the specified qubits

# implement the Quantum Fourier Transform (QFT)
## Add your Correct code Here (# define Number of qubits for QFT)
## Add your Correct code Here
n_qubits=3
qc=QuantumCircuit(n_qubits)
# Apply Quantum Fourier Transform (QFT) on the circuit
## Add your Correct code Here
apply_qft(qc,n_qubits)
# Draw the quantum circuit
print("Quantum Circuit with QFT:")
print(qc)
# Grading function to check if QFT has been applied to a quantum circuit
def grade_qft_application(qc, n_qubits):
    # Create an empty Quantum Circuit
    expected_qc = QuantumCircuit(n_qubits)

    # Apply QFT to the expected circuit
    apply_qft(expected_qc, n_qubits)

    # Compare the given circuit with the expected circuit
    if qc == expected_qc:
        print("Congratulations! QFT has been correctly applied.")
    else:
        print("Oops! QFT has not been correctly applied. Please review your implementation.")



# Call the grading function to check the QFT application
grade_qft_application(qc, n_qubits)
```
