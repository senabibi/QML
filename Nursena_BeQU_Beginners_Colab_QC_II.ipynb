## Exploring the Bell State Algorithm: Quantum Entanglement Magic

<p>We're about to dive into the intriguing realm of the Bell State Algorithm. This algorithm introduces us to the mysterious concept of quantum entanglement, often called "spooky action at a distance". Get ready for some quantum magic!</p>
    <h3>What is the Bell State Algorithm?</h3>
    <p>Bell States are a special family of quantum states that exhibit remarkable properties. They are like the "twins" of quantum states, and when we measure one, we instantly know something about the other, no matter how far apart they are. Bell States play a crucial role in various quantum applications, from quantum teleportation to quantum cryptography.</p>
    <h3>Quantum Entanglement: The Quantum Link</h3>
    <p>Before we dive into the Bell State algorithm, let's understand the concept of quantum entanglement. Imagine you have two qubits that are magically connected in a way that the state of one particles instantly influences the state of another. This mysterious connection is what we call entanglement. It's like having a pair of magical walkie-talkies that always communicate with each other, no matter the distance -- in formal terms, this is because the particles share a correlated quantum state.</p>
    <h3>The Bell State Algorithm</h3>
    <p>The Bell State algorithm is a simple but profound way to create Bell States. It involves two qubits, often referred to as Qubit A and Qubit B. Here's how it works:<p>
    <p>
    <ul>
        <li>Start with two qubits in the |0⟩ state.</li>
        <li>Apply a Hadamard gate to Qubit A.</li>
        <li>Apply a CNOT gate to Qubit B, using Qubit A as the control.</li></ul>
    </p>
    <h3>The Four Bell States</h3>
    <p>There are four Bell States, each with a unique quantum state. Here are the four Bell States:</p>
    <ol><li><strong>Bell State #1 (Φ+)</strong>: For input state |00⟩. They are in a superposition of |00⟩ and |11⟩.</li><li><strong>Bell State #2 (Φ-)</strong>: For input state |01⟩. They are in a superposition of |00⟩ and -|11⟩</li><li><strong>Bell State #3 (Ψ+)</strong>: For input state |10⟩. They are in a superposition of |01⟩ and |10⟩</li><li><strong>Bell State #4 (Ψ-)</strong>: For input state |11⟩. They are in a superposition of |01⟩ and -|10⟩</li></ol>
    <h3>Let's Create Some Bell States!</h3>
    <p>Now, let's roll up our sleeves and use Qiskit, a powerful quantum programming tool, to create these Bell states. We'll see the magic of entanglement in action!</p>

<h3>Bell State #1 (Φ+) circuit</h3>

1) Import the necessary packages.

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram, plot_state_qsphere
```

2) Let's build the Bell State #1 (Φ+) circuit.

```python
# We create a quantum circuit with 2 qubits and 2 classical bits
qc1 = QuantumCircuit(2, 2)

# We apply the Hadamard gate to the first qubit
qc1.h(0)

# We apply the CNOT gate to the first and second qubits
qc1.cx(0, 1)

# We draw the circuit
qc1.draw(output='mpl')
```

3) Let's see the result in the form of a histogram.

```python
# We execute the quantum circuit on a statevector simulator SamplerV2
from qiskit_aer.primitives import SamplerV2
from qiskit.visualization import plot_state_qsphere
sampler = SamplerV2()
job = sampler.run([qc1])
result = job.result()
state = Statevector(qc1)

# We plot the results
plot_state_qsphere(state)
```

`Here we see that the result is either |00⟩ or |11⟩. This is exactly what we expect from Bell State #1 (Φ+).`

## Unveiling the Bernstein-Vazirani Algorithm

<h3>Problem Statement</h3>
    <p>Imagine you have a oracle box that takes an input of 'n' bits and produces a single bit as output. This oracle box has a hidden 'n-bit secret string' that determines the output. Your task is to unveil this hidden string using the least number of queries to the oracle box.</p>
    <h3>The Classical Approach</h3>
    <p>Classically, to unveil the secret string, you would need to make 'n' queries to the black box, one for each bit of the secret string. This is known as the <strong>brute-force</strong> approach, and it takes 'n' queries, which is linear in the problem size.</p>
    <h3>Enter the Bernstein-Vazirani Algorithm</h3>
    <p>Now, let's introduce the quantum magic! The Bernstein-Vazirani algorithm is a quantum algorithm that can unveil the secret string in a single query, regardless of the problem size 'n'. That's right; it's exponentially faster than the classical approach.<p>
    <h3>How Does it Work?</h3>
    <p>The BV algorithm leverages quantum parallelism and the power of superposition to unveil the secret string efficiently. Here's how it works:</p>
    <ol><li>Prepare 'n' qubits in the |0⟩ state. Add |1⟩ state to the 'n+1'th qubit</li><li>Apply a Hadamard gate (H-gate) to all 'n+1' qubits. This puts them in a superposition of all possible bit strings.</li><li>Query the black box (the 'oracle') to obtain the output, which is a result of bitwise multiplication (inner product) of the secret string with the input.</li><li>Apply another round of H-gates to the first 'n' qubits.</li><li>Measure the first 'n' qubits. The measurement result will be the hidden secret string!</li></ol>
    <h3>Let's recap Bernstein-Vazirani Algorithm in 2 minute</h3>
    <p>Here's a quick recap of the BV algorithm in 2 minutes:</p>

    <h3>Let's Witness Quantum Speed</h3>
    <p>Now, let's dive into quantum coding and witness the BV algorithm in action using Qiskit, a powerful quantum computing framework. We'll create a quantum circuit to unveil the secret string using the BV algorithm.</p>

1) Let's import the necessary libraries and initialize the quantum simulator.

```python
#from HiddenCode import generate_hidden_binary_string
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
```

```python
import random

def generate_hidden_binary_string():
    # Generate a random length between 1 and 10
    length = random.randint(1, 10)

    # Generate a random binary string of the given length
    binary_string = ''.join(random.choices(['0', '1'], k=length))

    return binary_string
```

2) From HiddenCode we get the secret string and store it in a variable called `hidden_code`.

```python
# We check the length of the hidden binary string
hidden_code = generate_hidden_binary_string()
```

3) Let's measure the length of the secret string:

```python
length = len(hidden_code)
print("The length of the binary string is:", length)
```

4) Let's create the Bernstein-Vazirani quantum circuit according to the length of the secret string:

```python
# We define our quantum circuit
qc = QuantumCircuit(length + 1, length + 1)

# We apply X-gate to the last qubit
qc.x(length)
qc.barrier()

# We apply Hadamard gates to all qubits
for i in range(length + 1):
    qc.h(i)

qc.barrier()
# We apply the oracle box
for i in range(length):
    if hidden_code[i] == '1':
        qc.cx(i, length)
qc.barrier()

# We apply Hadamard gates to all qubits except the last one
for i in range(length):
    qc.h(i)


qc.barrier()

# We measure all qubits except the last one
for i in range(length):
    qc.measure(i, i)

# We draw the circuit
qc.draw(output='mpl')
```

5) Let's mesure the circuit:

```python
# We simulate the quantum circuit using the qasm simulator
from qiskit_aer.primitives import SamplerV2
sampler = SamplerV2()
job = sampler.run([qc], shots=1000)
result = job.result()
counts = result[0].data.c.get_counts()
```

6) Let's check if our circuit is correct or not:

```python
# We get the output of the first n+1 qubits
output = max(counts, key=counts.get)

# We delete the first bit from the output and reverse it
output = output[1:][::-1]

# We print the results
print("Hidden binary string:", hidden_code)
print("Output from the quantum circuit:", output)
plot_histogram(counts)
```

Here we can see that the result is correct. We have unveiled the secret string using the BV algorithm in a single query, which is exponentially faster than the classical approach.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import numpy as np
from qiskit_aer.noise import NoiseModel, depolarizing_error, ReadoutError
```

```python
def create_bell_state(state_type="|Phi+>"):
    qc = QuantumCircuit(2, 2)

    if state_type == "|Psi->":
        # Start with qubit 0 in state |1>
        qc.x(0)

    qc.h(0)           # Create superposition
    qc.cx(0, 1)       # Entangle qubits

    if state_type == "|Psi->":
        # Adjust for the negative phase and bit flip
        qc.z(0)
        qc.x(1)

    qc.measure([0, 1], [0, 1])
    return qc

# Test |Phi+>
qc_phi = create_bell_state("|Phi+>")
# Test |Psi->
qc_psi = create_bell_state("|Psi->")

simulator = AerSimulator()
result = simulator.run(qc_phi).result()
print("Counts for |Phi+>:", result.get_counts())
```

```python
def chsh_circuit(theta):
    qc = QuantumCircuit(2, 2)
    # 1. Create Bell State |Phi+>
    qc.h(0)
    qc.cx(0, 1)

    # 2. Rotate basis for one qubit to simulate CHSH test
    # We rotate qubit 1 by theta before measuring
    qc.ry(theta, 1)

    qc.measure([0, 1], [0, 1])
    return qc

# Running at a specific angle (e.g., pi/4) to see correlation
angle = np.pi/4
qc_chsh = chsh_circuit(angle)
counts = simulator.run(qc_chsh).result().get_counts()
print(f"Results at angle {angle:.2f}: {counts}")
```

```python
# Answer to Challenge 3
def bv_2bit(secret="11"):
    n = 2
    qc = QuantumCircuit(n + 1, n) # n input qubits + 1 ancilla

    # Initialization
    qc.x(n)       # Ancilla to |1>
    qc.h(range(n + 1))
    qc.barrier()

    # Oracle for "11"
    for i in range(len(secret)):
        if secret[i] == '1':
            qc.cx(i, n)
    qc.barrier()

    # Return to computational basis
    qc.h(range(n))
    qc.measure(range(n), range(n))
    return qc

qc_2bit = bv_2bit("11")
qc_2bit.draw('mpl')
```

```python
# Answer to Challenge 4
from qiskit_aer.noise import NoiseModel, depolarizing_error
from qiskit_aer import AerSimulator

# 1. Create the 3-bit BV Circuit
qc_noisy = QuantumCircuit(4, 3) # 3 bits + 1 ancilla
secret = "101"

qc_noisy.x(3)
qc_noisy.h(range(4))
# Oracle for 101
qc_noisy.cx(0, 3)
qc_noisy.cx(2, 3)
qc_noisy.h(range(3))
qc_noisy.measure(range(3), range(3))

# 2. Define a Noise Model (e.g., 5% error on every gate)
error_prob = 0.05
error = depolarizing_error(error_prob, 1) # 1-qubit gate error
noise_model = NoiseModel()
noise_model.add_all_qubit_quantum_error(error, ['h', 'x'])

# 3. Simulate with Noise
backend = AerSimulator(noise_model=noise_model)
result_noisy = backend.run(qc_noisy, shots=1024).result()
counts_noisy = result_noisy.get_counts()

print("Noisy results for '101':", counts_noisy)
# Note: You will see '101' as the highest peak, but other
# results (like '001' or '111') will appear due to noise.
```

```python
# Answer to Challenge 5
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, depolarizing_error, ReadoutError
from qiskit.visualization import plot_histogram

def run_noisy_bv_3bit(secret_string="101", error_rate=0.1):
    n = len(secret_string)
    qc = QuantumCircuit(n + 1, n)

    # --- BV Algorithm Steps ---
    qc.x(n)
    qc.h(range(n + 1))
    qc.barrier()

    # Oracle for '101'
    # Remember: Qiskit uses little-endian (bit 0 is on the right in strings)
    # If secret is '101', indices with '1' are 0 and 2.
    for i in range(n):
        if secret_string[::-1][i] == '1':
            qc.cx(i, n)

    qc.barrier()
    qc.h(range(n))
    qc.measure(range(n), range(n))

    # --- Noise Model Simulation ---
    # We add a depolarizing error (gate noise) and readout error (measurement noise)
    noise_model = NoiseModel()

    # 1-qubit gate error
    p_gate = error_rate
    error_gate = depolarizing_error(p_gate, 1)
    noise_model.add_all_qubit_quantum_error(error_gate, ['h', 'x'])

    # 2-qubit gate error (CNOT is usually noisier)
    error_cnot = depolarizing_error(p_gate * 2, 2)
    noise_model.add_all_qubit_quantum_error(error_cnot, ['cx'])

    # --- Execution ---
    simulator = AerSimulator(noise_model=noise_model)
    job = simulator.run(qc, shots=1024)
    counts = job.result().get_counts()

    return counts, qc

counts, circuit = run_noisy_bv_3bit("101", 0.05)
print("Counts with noise:", counts)
plot_histogram(counts)
```
