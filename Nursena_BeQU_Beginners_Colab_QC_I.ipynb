## X-Gate (Bit Flip)

<p>Think of the X-gate as a magic switch for your quantum computer. It's like flipping a light switch – it changes things from one state to another. Let's see how it works in a fun way:</p>
<h3>What is the X-Gate (Bit Flip)?</h3>
<p>The X-gate is a bit like a trickster. If your qubit is in the "OFF" state, represented as ∣0⟩, the X-gate will quickly turn it "ON", making it ∣1⟩. On the other hand, if your qubit is already "ON" as ∣1⟩, the X-gate will play another trick and turn it "OFF" to ∣0⟩.</p>
<h3>Imagine a Toggle Switch</h3>
<p>Imagine your qubit as a tiny lightbulb. If it's off, the X-gate flips the switch, and the light turns on. If it's already on, the X-gate flips it again, and the light turns off. It's like having a remote control for your quantum light!</p>

<h3>How to use the X-Gate?</h3>
<p>Let's see how the X-gate works in Qiskit.</p>

1) We need to import the necessary packages:

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
from qiskit.visualization import circuit_drawer
from qiskit_aer.primitives import SamplerV2
```

2) We set up a simple circuit with one qubit, and one classical bit to measure the qubit with:

```python
# Define the number of quantum bits in the circuit
quantum_bit = 1

# Define the number of classical bits in the circuit
classical_bit = 1

# Create a quantum circuit with the defined number of quantum and classical bits
circ = QuantumCircuit(quantum_bit, classical_bit)
```

In this code:

- `quantum_bit = 1` sets the number of quantum bits (qubits) in the quantum circuit. A quantum bit, or qubit, is the fundamental unit of quantum information.

- `classical_bit = 1` sets the number of classical bits in the quantum circuit. A classical bit is the traditional binary bit used in classical computing, which can be either 0 or 1.

- `circ = QuantumCircuit(quantum_bit, classical_bit)` creates a new quantum circuit. A quantum circuit is a sequence of quantum gates, and it's the basic unit of quantum computation.

3) We add the X-gate to our circuit:

```python
circ.x(0)
```

4) Let's see the circuit:

```python
circuit_drawer(circ, output="mpl", style={"backgroundcolor": "#EEEEEE"})
```

5) Let's see the result of the circuit in a bloch sphere:

```python
state = Statevector(circ)
plot_bloch_multivector(state)
```

6) We measure it:

```python
# Measure the quantum bit and store the result in the classical bit
circ.measure(0, 0)
```

7) And now, let's run the circuit 1000 times, and plot the results:

```python
from qiskit.providers.basic_provider import BasicProvider
backend = BasicProvider().get_backend('basic_simulator')
sampler = SamplerV2()
job = sampler.run([circ], shots=1000)
result = job.result()
counts = result[0].data.c.get_counts()
plot_histogram(counts)
```

As you can see, the result is always 1, meaning that the qubit is always in the ∣1⟩ state. This is because the X-gate flips the qubit from ∣0⟩ to ∣1⟩.

8) Let's try it again, but this time, we start with the qubit in the ∣1⟩ state:

```python
# Create a new quantum circuit with one quantum bit (qubit) and one classical bit
circ = QuantumCircuit(1, 1)

# Apply an X gate (also known as a NOT gate) to the qubit at index 0
# The X gate flips the state of the qubit from |0> to |1> and vice versa
circ.x(0)

# Apply another X gate to the qubit at index 0
# Since the X gate is its own inverse, applying it twice returns the qubit to its original state
circ.x(0)

# Draw the quantum circuit
# The 'output' parameter is set to 'mpl' to use Matplotlib for drawing
# Each horizontal line represents a qubit, and time flows from left to right
# The boxes represent quantum gates, and the double lines represent measurements
circuit_drawer(circ, output="mpl", style={"backgroundcolor": "#EEEEEE"})
```

9) Let's see the result of the circuit in a bloch sphere:

```python
state = Statevector(circ)
plot_bloch_multivector(state)
```

10) We measure it:

```python
circ.measure(0, 0)
```

11) And now, let's run the circuit 1000 times, and plot the results:

```python
sampler = SamplerV2()
job = sampler.run([circ], shots=1000)
result = job.result()
counts = result[0].data.c.get_counts()
plot_histogram(counts)
```

As you can see, the result is always 0, meaning that the qubit is always in the ∣0⟩ state. This is because the X-gate flips the qubit from ∣1⟩ to ∣0⟩. So, the X-gate is a bit like a light switch. If the light is off, it turns it on. If the light is on, it turns it off.

## The Hadamard Gate (H gate)

<p>In the enchanting world of quantum computing, the Hadamard gate (often called the H gate) is like a magician's wand, performing a mesmerizing trick with quantum bits or "qubits". This gate is one of the fundamental tools that give quantum computers their superpowers. In this section, we'll break down the Hadamard gate in simple terms.</p>
    <h3>What is the Hadamard Gate?</h3>
    <p>Imagine you have a quantum bit, or "qubit". A regular computer bit can be either a 0 or a 1, like a light switch being off or on. But the H gate is a special tool that allows qubits to do something extraordinary: it puts them in a state of "superposition". But what's superposition? Imagine it as a bit like Schrödinger's cat being both alive and dead at the same time until you look. In quantum terms, it means that the qubit can be in multiple states simultaneously. It's like having a spinning coin that's both heads and tails at the same time.</p>
    <h3>How Does the H Gate Work?</h3>
    <p>Here's the magic of the H gate:</p>
    <ol>
        <li>If you start with a qubit in the state |0⟩ (like a light switch off), applying the H gate turns it into a blend of |0⟩ and |1⟩. It's as if the switch is flickering between on and off!</li>
        <br>
        <br>

        <br>
        <br>
        <li>If you begin with a qubit in the state |1⟩ (like a light switch on), guess what? The H gate does the same thing, making it a mix of |0⟩ and <strong>-|1⟩</strong>. Now, it's like the switch is winking on and off!</li>

    <p> The H gate is a 90° rotation around the X+Z axis on the Bloch sphere. It's a fundamental tool in quantum computing, and it's used in many quantum algorithms. It's also the key to understanding the Bell state, which is the basis of quantum entanglement.</p>

<h2>What if we apply an H gate, then measure it, and apply an H gate again?</h2>
<p>Let's perform a quantum experiment that involves applying an H (Hadamard) gate, measuring, and then applying another H gate. This experiment demonstrates the probabilistic and reversible nature of quantum operations. Follow along step by step:</p>

1) Apply H-Gate

```python
from qiskit.visualization import plot_bloch_multivector
qc4 = QuantumCircuit(1,1)
qc4.h(0)
```

2) Measure the qubit

```python
qc4.measure(0,0)
```

3) Apply H-Gate again

```python
qc4.h(0)
```

4) Draw the circuit

```python
circuit_drawer(qc4, output="mpl", style={"backgroundcolor": "#EEEEEE"})
```

5) See the results using a bloch sphere:

```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
from qiskit.primitives.statevector_sampler import StatevectorSampler
qc4 = QuantumCircuit(1,1)
qc4.h(0)
qc4.h(0)
sampler = SamplerV2()
job = sampler.run([qc4], shots=1024)
result = job.result()
state = Statevector(qc4)
plot_bloch_multivector(state)
```

*See\*, each time we are runnig the `step 5`, we are getting either ∣+⟩ or ∣-⟩ state. But we applied H gate twice. So, we should get the initial state of the qubit, right? But why it's returning ∣+⟩ or ∣-⟩ state with 50% probability? Good question!

<h2>Why?</h2>
    <p>Let's see what happed there:</p>
    <h2>Applying Two H-Gates Without Measurement in Between</h2>
    <p>When you apply an H-gate twice in a row without any measurement in between, here's what happens:</p>
    <ol>
        <li>Start with a qubit in the state |0⟩</li>
        <li>Apply the first H-gate, which transforms it into |+⟩ </li>
        <li>Apply the second H-gate, which transforms it into |0⟩ again. So, applying two H-gates in a row without measurement restores the qubit to its original state ∣0⟩. This demonstrates the reversible nature of quantum gates.</li>
    </ol>
    <h2>Applying Two H-Gates With a Measurement in Between</h2>
    <p>When you apply an H-gate twice in a row with a measurement in between, here's what happens:</p>
    <ol>
        <li>Start with a qubit in the state |0⟩</li>
        <li>Apply the first H-gate, which transforms it into |+⟩ </li>
        <li>Measure the qubit. It now has a 50% chance of being in either state |0⟩ or |1⟩.</li>
        <li>Let's say the result is |1⟩. Apply the second H-gate, which transforms it into |−⟩ </li>
        <li>Measure the qubit. It now has a 50% chance of being in either state |0⟩ or |1⟩.</li>
        <li>Let's say the result is |0⟩. Apply the second H-gate, which transforms it into |+⟩</li>
    </ol>
    <p>So, the crucial distinction is that measurement in between the two H-gates introduces uncertainty in the qubit's state, whereas without measurement, the qubit returns to its original state deterministically. This difference is a fundamental aspect of quantum computation, where measurements can alter the outcomes of subsequent operations.</p>

## The CX (CNOT) Gate: A Key Player in Quantum Computing

<p>In this notebook, we'll take a gentle and easy-to-understand journey into one of the most essential quantum gates: the CX gate, also known as the CNOT gate.</p>
<h3>What is the CX Gate?</h3>
<p>The CX gate is a building block of quantum circuits. It's like a special switch that can change the state of one qubit (quantum bit) based on the state of another qubit. We have two qubits involved: a control qubit and a target qubit.</p>
    <ol>
        <li><strong>Control Qubit</strong>: Think of this as the "boss" qubit.</li>
        <br>
        <br>
            <center>
              <img src="https://drive.google.com/uc?id=1HvqPP1LDH8LneH_XTXHtZ4pwLfIs4Fvs" alt="CxGate" width="600">
            </center>
        <br>
        <br>
        <li><strong>Target Qubit</strong>: This is the "follower" qubit.</li>
    </ol>
    <h3>How Does the CX Gate Work?</h3>

## Exploring the Z Gate: Unveiling Quantum Phase

<h3>What is the Z Gate?</h3>
    <p>Imagine you have a magical tool that can change the way a qubit thinks. That tool is the Z gate. When we apply the Z gate to a qubit, it doesn't flip it like the X gate. Instead, it tweaks the phase of the qubit's state.</p>
    <ul><li>If the qubit is in state |0⟩, the Z gate leaves it unchanged.</li><li>If the qubit is in state |1⟩, the Z gate adds a phase (a twist) to it.</li></ul>
    <h3>The Math Behind the Z Gate (Don't Worry, It's Simple)</h3>
    <p>Okay, we'll touch a tiny bit on the math, but don't be scared! The Z gate applies a phase shift of π (pi) to the |1⟩ state. So, mathematically, if |ψ⟩ is our qubit's state:</p>
    <ul><li>Z(|0⟩) = |0⟩ (no change)</li><li>Z(|1⟩) = -|1⟩ (adds a phase of π)</li></ul>
    <p>You don't need to remember the math, but it's good to know the Z gate does this twisty thing.</p>
